```python
# Remove rows with any missing values
new_retail_data_cleaned = new_retail_data.dropna()

# Display the number of missing values per column after removal
print("\nMissing values per column after removal:")
print(new_retail_data_cleaned.isnull().sum())

# Display the shape of the dataset before and after removal
print("\nShape of dataset before removal:", new_retail_data.shape)
print("Shape of dataset after removal:", new_retail_data_cleaned.shape)
```

Geographics

```python
country_info = new_retail_data_cleaned[['Country']]
country_info

# Count the occurrences of each country
country_counts = new_retail_data_cleaned['Country'].value_counts()


# Plot the pie chart
plt.figure(figsize=(10, 6))
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=140, colors=custom_colors)
plt.title('Distribution of Transactions by Country')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

Age

```python
# Define the function to categorize ages into generations
def categorize_generation(age):
    if 11 <= age <= 26:
        return 'Generation Z'
    elif 27 <= age <= 42:
        return 'Millennials'
    elif 43 <= age <= 58:
        return 'Generation X'
    elif 59 <= age <= 77:
        return 'Baby Boomers'
    else:
        return 'Other'

# Apply the function to the Age column
new_retail_data_cleaned['Generation'] = new_retail_data_cleaned['Age'].apply(categorize_generation)

# Display the first few rows to check the new column
print(new_retail_data_cleaned[['Age', 'Generation']].head())

# Optionally, you can group by the Generation and count the number of occurrences
generation_counts = new_retail_data_cleaned['Generation'].value_counts()

# Display the counts of each generation
print(generation_counts)
```

```python
# Plot the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(generation_counts.index, generation_counts.values, color=custom_colors)

# Add numbers above the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom')  # va: vertical alignment

plt.title('Distribution of Customers by Generation')
plt.xlabel('Generation')
plt.ylabel('Number of Customers')
plt.xticks(rotation=45)
plt.show()
```

###*Sales Trend*

Through out the years

```python
# Create a Month_Year column
new_retail_data_cleaned['Month_Year'] = new_retail_data_cleaned['Date'].dt.to_period('M')

# Group by Month_Year and Product_Category, summing the sales amounts
sales_trend = new_retail_data_cleaned.groupby(['Month_Year', 'Product_Category'])['Total_Amount'].sum().unstack()

# Plotting
plt.figure(figsize=(12, 6))

# Plot each product category's sales trend
for category in sales_trend.columns:
    plt.plot(sales_trend.index.astype(str), sales_trend[category], marker='o', linestyle='-', label=category)

# Add labels, title, and legend
plt.title('Sales Trend Over Time by Product Category', fontsize=16)
plt.xlabel('Month and Year', fontsize=14)
plt.ylabel('Total Amount', fontsize=14)
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)
plt.legend(title='Product Category', bbox_to_anchor=(1.05, 1), loc='upper left')  # Add legend outside of the plot

# Display the plot
plt.tight_layout()  # Adjust layout to prevent clipping of tick-labels
plt.show()
```

Month

```python
# Define the correct order of months
month_order = ['January', 'February', 'March', 'April', 'May', 'June','July', 'August', 'September', 'October', 'November', 'December']

# Reindex the sales_trend_monthly to ensure it follows the month order
'''
Bach:
Not sure what you meant sales_trend_monthly. I add the 2 following code lines as similar to the weekly.
'''
new_retail_data_cleaned['Month_name'] = new_retail_data_cleaned['Date'].dt.month_name()
sales_trend_monthly = new_retail_data_cleaned.groupby('Month_name')['Total_Amount'].sum()


sales_trend_monthly.index = pd.Categorical(sales_trend_monthly.index, categories=month_order, ordered=True)
sales_trend_monthly = sales_trend_monthly.sort_index()

# Create the bar graph
plt.figure(figsize=(10, 6))
bars = sales_trend_monthly.plot(kind='bar', color=custom_colors)

# Add titles and labels
plt.title('Monthly Sales Trend', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Total Sales Amount', fontsize=14)

# Show the grid for better readability
plt.grid(axis='y')

# Display the graph
plt.xticks(rotation=45)

# Add numbers on top of the bars
for bar in bars.patches:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 50,  # Offset the number a little above the bar
        int(bar.get_height()),  # Get the height of the bar (sales amount)
        ha='center',  # Center align the text
        fontsize=10,  # Set font size
        color='black'  # Set font color
    )

plt.tight_layout()
plt.show()
```

Day

```python
# Extract the day name from the Date
new_retail_data_cleaned['Day_name'] = new_retail_data_cleaned['Date'].dt.day_name()

# Define the correct order of the week
week_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

# Aggregate sales by day of the week
sales_trend_weekly = new_retail_data_cleaned.groupby('Day_name')['Total_Amount'].sum()

# Reindex sales_trend_weekly to ensure it follows the week order
sales_trend_weekly.index = pd.Categorical(sales_trend_weekly.index, categories=week_order, ordered=True)
sales_trend_weekly = sales_trend_weekly.sort_index()

# Define custom colors for the bars (modify as needed)
custom_colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2']

# Create the bar graph
plt.figure(figsize=(10, 6))
bars = sales_trend_weekly.plot(kind='bar', color=custom_colors)

# Add titles and labels
plt.title('Weekly Sales Trend', fontsize=16)
plt.xlabel('Day of the Week', fontsize=14)
plt.ylabel('Total Sales Amount', fontsize=14)

# Show the grid for better readability
plt.grid(axis='y')

# Display the graph
plt.xticks(rotation=45)

# Add numbers on top of the bars
for bar in bars.patches:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 20,  # Offset the number a little above the bar
        int(bar.get_height()),  # Get the height of the bar (sales amount)
        ha='center',  # Center align the text
        fontsize=10,  # Set font size
        color='black'  # Set font color
    )

plt.tight_layout()
plt.show()
```
